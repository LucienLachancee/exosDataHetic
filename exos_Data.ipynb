```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f2(x):
    return np.exp(-x/10)*np.sin(x)

```

Question 1 et 2 :
CF [0,10] sur 10 000 points
CF' [0,10] la dérivée

```python
a, b = 0, 10
nmb_points = 10000

listex = np.linspace(a,b, nmb_points)
listey= f2(listex)

df_dx= np.gradient(listey,listex)

print(listex)
```

```python
plt.figure(figsize=(15,6))
plt.plot(listex, listey)
plt.plot(listex, df_dx)
plt.show()
```

Question 3:

Faire la moyenne et l'écart-type de F sur [3,7]

```python
import statistics

moy  = np.mean(listey[3:7])

test = np.std(listey[3:7])
test2 = statistics.stdev(listey[3:7])


print(moy,test,test2)
```

```python
sum = 0
count = 0
for i in range(3, 7):
    sum += listey[i]
    count += 1
moyenne = sum / count

variance = 0
for i in range(3, 7):
    variance += (listey[i] - moyenne)**2


ecart_type = (variance / (count - 1))**0.5

print( moyenne,ecart_type)
```

Question 5:
Algorithme de Monte Carlo

```python
import random


c, d = 0, 10
nmb_p = 10000
total = 0
for i in range(nmb_p):
    x = random.uniform(c, d)  # Mettre plein de points aléatoire entre c et d
    total += f2(x)  # total = calculer sur l'instant t la fonction par rapport à l'échantillon

mean = total/nmb_p  #mean = step
integral = mean*(d-c)  

print(integral)
```
